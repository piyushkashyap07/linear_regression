```python
import numpy as np
```

```python
trainingSet=np.loadtxt(r'C:\Piyush\Scripts\Machine-Learning\Linear_regression\0000000000002329_training_diabetes_x_y_train.csv',delimiter=",")
testingSet=np.loadtxt(r'C:\Piyush\Scripts\Machine-Learning\Linear_regression\0000000000002329_test_diabetes_x_test (1).csv',delimiter=",")
```

```python
x=trainingSet[:,:10]
y=trainingSet[:,-1]
```

```python
from sklearn import model_selection
x_train,x_test,y_train,y_test=model_selection.train_test_split(x,y)
```

```python
from sklearn.linear_model import LinearRegression
alg1=LinearRegression()
```

```python
alg1.fit(x_train,y_train)
```

```python
y_predict=alg1.predict(x_test)
```

```python
import matplotlib.pyplot as plt
plt.scatter(y_test,y_predict)
plt.axis([0,300,0,300])
plt.show()
```

```python
y_Test_prediction=alg1.predict(testingSet)
```
